```python
import pandas as pd
import requests
from html.parser import HTMLParser
import functools
from bs4 import BeautifulSoup
import csv
import sys
from pylab import rcParams
import seaborn as sns
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
import statsmodels.api as sm
import datetime
import calendar
```

```python
#Лучше не запускать, работает 20 минут. Данные сохранены в  fanfics.csv 


all_fanfics=[]
for i in range(917):
    res=[]
    url = "https://fanfics.me/fandom2/fics?page="+str(i+1)
    #print(url)
    html_1 = requests.get(url).text
    #print(html_1)
    soup = BeautifulSoup(html_1, 'html.parser')

    fanfics=soup.find_all(class_="FicTable")
    res=[{"Views": int(functools.reduce(lambda s1,s2: s1+s2,fanfic.find(class_="Views")["title"].split()[2:])),
          "Readers":int(functools.reduce(lambda s1,s2: s1+s2,fanfic.find(class_="ReadersCount")["title"].split()[2:])),
          "Comments": int(functools.reduce(lambda s1,s2: s1+s2,fanfic.find(class_="Comments")["title"].split()[2:])),
          "Recommends":int(fanfic.find(class_="RecommendsCount").text),
          "StartWriting":fanfic.find(class_="DateUpdate")["title"].split()[1][:-1],
          "LastUpdate": fanfic.find(class_="DateUpdate")["title"].split()[3],
          "Status":fanfic.find(class_="FicTbl_meta").find(class_="FicTable_Status").text,
          "Size":int(fanfic.find(class_="FicTbl_meta").find(class_="FicTable_Size").text.split()[2]),
          "SizeType":fanfic.find(class_="FicTbl_meta").find(class_="FicTable_Size").text.split()[0],
          "Author":fanfic.find(class_="FicTbl_meta").find(class_="user").text,
          "FanficsTitle":fanfic.find(class_="FicTable_Title").find("h4").find("a").text,
          "FanficsType":fanfic.find(class_="FicTable_Title").find("h4").find("span").text[1:-1],
         } 
        for fanfic in fanfics
         if fanfic.find(class_="FicTbl_meta") and fanfic.find(class_="FicTbl_meta").find(class_="user")
        
        ]
    all_fanfics+=res
#print(all_fanfics)


#for fanfic in fanfics:
    #print(fanfic.find(class_="DateUpdate").text)
    #print(fanfic.find(class_="Comments").text)
    #print(fanfic.find("span",class_="ReadersCount"))
```

Остановимся подробнее на описании информации, полученной после парсинга сайта:

- Views - количеству просмотров фанфика (натуральное число) 
- ReadersCount - число подписчиков фанфика (натуральное число) 
- Comments - количеству комментариев к фанфику (натуральное число) 
- Recommends - сколько пользоватейлей написало отзыв на фанфик (натуральное число)
- StartWriting - дата начала публикации работы 
- LastUpdate - дата последнего обновления фанфика (Фанфики обычно выкладываются главами, поэтому два последних столбца могут сильно отличаться)
- Status - есть три типа фанфика, по его текущему состоянию:
    - Закончен - работа дописана
    - В процессе - работа публикуется, но еще не дописана до конца
    - Заморожен - автор отложил написание фанфика на неопределенное время
- Size - размер фанфика в Килобайтах (натуральное число)
- SizeType - градация фанфика по размеру:
    - Мини - примерно сооответствует рассказу
    - Миди - примерно соответствует повести
    - Макси -примерно соответствует роману
- Author - никнэйм автора фанфика
- FanficsTitle -название фанфика
- FanficsType - джен, гет, слэш или фемслэш.

Получили то, что и предполагали: выборки просмотров не распределены нормально. Это лишает нас возможности исползовать параметрические критерии.

Проверим, сначала гипотезу H0 о том, что среднее число просмотров фанфика не зависит от того, к какой категории он относится. Для этого воспользуемся критерием $\chi^2$. В таблицу истинности поставим значения среднего по всей выборке, а в таблицу значений-средние по подвыборкам из Макси, Миди и Мини фанфиков соответсвенно.

```python
real_means=[list_fanfics[list_fanfics["SizeType"]=="Мини"]["Views"].mean(),
            list_fanfics[list_fanfics["SizeType"]=="Миди"]["Views"].mean(),
            list_fanfics[list_fanfics["SizeType"]=="Макси"]["Views"].mean()]
gip_means=[list_fanfics["Views"].mean()]*3
print(real_means,"\n",gip_means)
chisquare(real_means,gip_means)
```

Отсюда видно, что средние по подвыборкам отличаются черезвычайно сильно: на много порядков ниже уровня статистической значимости. 

С помощью критерия Колмогорова-Смирнова проверим гипотезы о том, что $F_{mini}< F_{midi} < F_{maxi}$

```python
print(stats.ks_2samp(list_fanfics[list_fanfics["SizeType"]=="Мини"]["Views"].values,
             list_fanfics[list_fanfics["SizeType"]=="Миди"]["Views"].values,alternative="greater"))
print(stats.ks_2samp(list_fanfics[list_fanfics["SizeType"]=="Миди"]["Views"].values,
             list_fanfics[list_fanfics["SizeType"]=="Макси"]["Views"].values,alternative="greater"))
```

Таким образом мы принимаем альтернативу $F_{mini}\geq F_{midi} \geq F_{maxi}$. Этому можно найти также графическое подтверждение-построив графики эмпирических функций распределения по выборкам.

```python
#эмпирическая функция распределения
def edf_l(l,x):
    return np.searchsorted(l, x, side='right') / l.size
```

```python

l_maxi=np.sort(list_fanfics[list_fanfics["SizeType"]=="Макси"]["Views"])
l_midi=np.sort(list_fanfics[list_fanfics["SizeType"]=="Миди"]["Views"])
l_mini=np.sort(list_fanfics[list_fanfics["SizeType"]=="Мини"]["Views"])
xvalue=np.arange(0,300000,100)
yvalue = [edf_l(l_maxi,x) for x in xvalue]
plt.plot(xvalue,yvalue,color="blue")
yvalue = [edf_l(l_midi,x) for x in xvalue]
plt.plot(xvalue,yvalue,color="red")
yvalue = [edf_l(l_mini,x) for x in xvalue]
plt.plot(xvalue,yvalue,color="green")
```

Таким образом Макси-фанфики просматривают больше чем Миди, а Миди-чем Мини. Если проводить анализ по другим признакам: числу просмотров, комментариев, реккомендаций, то получим примерно то же самое. Можно сказать, что в Макси-фанфики любят и читают больше. 


```python
l_maxi=np.sort(list_fanfics[list_fanfics["SizeType"]=="Макси"]["Recommends"])
l_midi=np.sort(list_fanfics[list_fanfics["SizeType"]=="Миди"]["Recommends"])
l_mini=np.sort(list_fanfics[list_fanfics["SizeType"]=="Мини"]["Recommends"])
xvalue=np.arange(0,100,1)
yvalue = [edf_l(l_maxi,x) for x in xvalue]
plt.plot(xvalue,yvalue,color="blue")
yvalue = [edf_l(l_midi,x) for x in xvalue]
plt.plot(xvalue,yvalue,color="red")
yvalue = [edf_l(l_mini,x) for x in xvalue]
plt.plot(xvalue,yvalue,color="green")
```

Построим предказание временного ряда-хотим узнать, сколько фанфиков было опубликовано 1 декабря (в реальности их было 3).

Построим предсказание по взвешенному среднему, опираясь, в основном на трейн.

```python
np.average(datas[-365:])
```

Как видим, результат не соответствует действительности. Построим предсказание используя экспотенциальное сгляживание.

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # first value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result[-1]
```

```python
exponential_smoothing(datas.values, 0.2)
```

Уже ближе. А теперь двойное эспотенциальное сглаживание.

```python
def double_exponential_smoothing(series, alpha, beta):
    result = [series[0]]
    for n in range(1, len(series)+1):
        if n == 1:
            level, trend = series[0], series[1] - series[0]
        if n >= len(series): # прогнозируем
            value = result[-1]
        else:
            value = series[n]
        last_level, level = level, alpha*value + (1-alpha)*(level+trend)
        trend = beta*(level-last_level) + (1-beta)*trend
        result.append(level+trend)
    return result[-1]
```

```python
double_exponential_smoothing(datas.values, 0.2,0.73)
```
